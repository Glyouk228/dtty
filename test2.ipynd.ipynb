```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
f = float(input('Введите значение для переменной f'))
y = float(input('Введите значение для переменной y'))

print(f'G ={np.exp(2 * y) + np.sin(f) / np.log(3.8 * y + f)}')
```

```python
class MyClassTask1:
    def __init__(self, f, y):
        self.f = f
        self.y = y

    def calculate_G(self):
        print(f'G ={np.exp(2 * y) + np.sin(f) / np.log(3.8 * y + f)}')
```

```python
f = float(input('Введите значение для переменной f'))
y = float(input('Введите значение для переменной y'))

my_class = MyClassTask1(f,y)
my_class.calculate_G()
```

```python
y_values = np.linspace(0.1, 10, 100)
G_values = np.e * 2 * y_values + np.sin(f) / np.log(3.8 * y_values + f)

plt.plot(y_values, G_values, label='G(y)')
plt.xlabel('y')
plt.ylabel('G')
plt.title('График G как функция от y')
plt.axhline(0, color='black', linewidth=0.5, ls="--")
plt.axvline(0, color='black', linewidth=0.5, ls="--")
plt.grid()
plt.legend()
plt.show()
```
